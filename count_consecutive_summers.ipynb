## Problem statement

Positive integers can be expressed as sums of consecutive positive integers in various ways. For example, 42 can be expressed as such a sum in four different ways:
1. 3+4+5+6+7+8+9, 
2. 9+10+11+12, 
3. 13+14+15 and 
4. 42. 

As the last solution (d) shows, any positive integer can always be trivially expressed as a singleton sum that consists of that integer alone.

Compute how many different ways it can be expressed as a sum of consecutive positive integers.

Input: Int

Output: Int

Example:
```
	count_consecutive_summers(42) == 4
	count_consecutive_summers(99) == 6
```

### Precondition: 
Input is always a positive integer.

````
# this solution splits a uniform median into odd and even cases
def count_consecutive_summers(num):
    summer_count = 0
    print(num)
    for factor in range(1, num + 1):
        factor_complement = float(num) / float(factor)
        if (factor_complement - (factor / 2) > 0):
            if ((factor % 2 == 1) and (factor_complement - int(factor_complement) == 0)):
                summer_count = summer_count + 1
            if ((factor % 2 == 0) and (factor_complement - int(factor_complement) == 0.5)):
                summer_count = summer_count + 1
    return summer_count
            


if __name__ == '__main__':
    print("Example:")
    print(count_consecutive_summers(42))

    # These "asserts" are used for self-checking and not for an auto-testing
    '''
    assert count_consecutive_summers(42) == 4
    assert count_consecutive_summers(99) == 6
    assert count_consecutive_summers(1) == 1
    '''
    count_consecutive_summers(42)
    count_consecutive_summers(99)
    count_consecutive_summers(1)
    
    print("Coding complete? Click 'Check' to earn cool rewards!")
````
